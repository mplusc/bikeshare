```python
#Importing Libraries
import pandas as pd
import numpy as np
import csv
```

```python
#Assigning data files to pandas dataframe
Jan = pd.read_csv('static/data/201801_capitalbikeshare_tripdata.csv')
Jul = pd.read_csv('static/data/201807-capitalbikeshare-tripdata.csv')
```

```python
#Preview dataset of January Washington DC Capital Bikeshare
Jan.head(5)
```

```python
#Preview dataset of July Washington DC Capital Bikeshare
Jul.head(5)
```

```python
#Converting the duration to minutes
Jan['Duration'] = Jan['Duration']/60
Jul['Duration'] = Jul['Duration']/60
#Round Values
Jan['Duration'] = Jan['Duration'].round(decimals=2)
Jul['Duration'] = Jul['Duration'].round(decimals=2)
```

```python
Jan.head(5)
```

```python
#Formatting start date
Jan['Start date'] = pd.to_datetime(Jan['Start date'])
Jul['Start date'] = pd.to_datetime(Jul['Start date'])
```

```python
Jan.head(1)
```

```python
#Adding the Days of the week and time to the data frame
Jan['Day'] = Jan['Start date'].dt.weekday_name
Jan['Time'] = Jan['Start date'].dt.time
Jul['Day'] = Jul['Start date'].dt.weekday_name
Jul['Time'] = Jul['Start date'].dt.time
```

```python
Jan.head(2)
```

```python
Jan.head(5)
```

```python
#Dropping unecessary columns
Jan.drop(['Start date','End date','Bike number'], axis = 1, inplace = True)
Jul.drop(['Start date','End date','Bike number'], axis = 1, inplace = True)
```

```python
Jan.head(5)
```

```python
#output to csv
Jan.to_csv(r'static\data\clean_data\Jan_2018_DC_CaBi.csv', encoding = 'utf-8', index = False, columns = ['Duration', 'Start station number','Start station','End station number','End station','Member type','Day','Time'])
#output to csv
Jul.to_csv(r'static\data\clean_data\Jul_2018_DC_CaBi.csv', encoding = 'utf-8', index = False, columns = ['Duration', 'Start station number','Start station','End station number','End station','Member type','Day','Time'])
```
